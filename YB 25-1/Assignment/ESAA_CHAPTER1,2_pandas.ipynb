## 1-4. 데이터 핸들링 - 판다스
>  **판다스**: 2차원 데이터를 효율적으로 가공/처리하기에 유용한 파이썬 라이브러리.

* 리스트, 컬렉션, 넘파이, CSV 등의 파일을 쉽게 데이터프레임으로 변경
* 다양한 포맷으로 구성된 파일을 데이터프레임으로 로딩 가능.
* Index: 개별 데이터를 고유하게 식별하는 key 값.
* series: 칼럼 1개 cf) 데이터프레임 - 칼럼 여러개
* read_csv: 어떤 필드 구분 문자 기반의 파일 포맷 가능 / sep='' 으로 지정 / 디폴트 - ',' / read_table: 구분 문자 tab / read_fwf: 고정 길이 기반의 칼럼 포맷을 로딩
* read_cvs : 파일의 맨 처음 로우를 칼럼명으로 인지, 변환
* 모든 데이터프레임 내의 데이터는 생성되는 순간 고유의 인덱스값 가지게 됨
* shape 로 행, 열 크기 확인
* *info()* : 총 데이터 건수, 데이터 타입, NULL 건수 확인
* *describe()* : 칼럼별 숫자형(only) 데이터값의 percentile 분포도, 평균값, 최댓값, 최솟값 계산
  - 숫자형 카테고리 칼럼 판단에 이용
* *value_count()* : series 칼럼에 적용 > 해당 칼럼의 유형과 건수 확인 가능
  - Null 값 제외하고 결과 반환 / droupna=True (디폴트)  
* 인덱스: 고유성 보장 필수 / 문자열도 할당 가능 / 의미있는 데이터값 할당 가능


> **데이터프레임과 리스트, 딕셔너리, 넘파이 ndarray 상호변환**

(1) 넘파이 ndarray, 리스트, 딕셔너리 -> 데이터프레임
* 2차원 이하의 데이터들만 변환 가능
* data= :변환하려는 객체 입력
* columns= : 지정하려는 칼럼명 리스트 객체 입력
* [딕셔너리 변환] key 는 칼럼명으로, value는 데이터값으로 변환됨

(2) 데이터프레임 -> 넘파이 ndarray, 리스트, 딕셔너리
* *.values* : ndarray 로 쉽게 변환!
* *.tolist()* : 리스트로 변환
* *.to_dict()* : 딕셔너리로 변환

> **데이터프레임의 칼럼 데이터 세트 생성과 수정**

(1) 칼럼 데이터 추가 - 일괄적으로 값 할당
* [] 내에 새로운 칼럼명 입력 후, 값 할당 >> 모든 값이 해당값으로 할당된,새로운 series가 기존 df 에 추가

(2) 칼럼 데이터 추가 - 기존 칼럼 series 이용

(3) 기존 칼럼 수정(업데이트)
* [] 내에 수정하고 싶은 칼럼명 입력 후, 값 할당

> **데이터프레임 데이터 삭제**

* *drop()* 사용
* [labels & axis 인자] axis 값에 따라 특정 칼럼 또는 특정 행 드롭
   - axis=0 - 특정 로우를 드롭 & labels - 인덱스값
   - axis=1 - 특정 칼럼을 드롭 & labels - 칼럼명

* [inplace 인자] 디폴트: inplace=False
    - 원본 데이터프레임의 데이터 삭제 X
    - 드롭된 데이터프레임을 새로운 객체 변수로 반환
    - inplace=True : 원본 데이터프레임의 데이터 삭제 / 반환 값이 None  (반환 값을 다시 원래의 데이터프레임 객체에 할당하는 것 주의)

> **Index 객체**

* 데이터프레임, series의 레코드를 고유하게 식별하는 객체.
* .index : 인덱스 객체만 추출 / 반환객체는 1차원 ndarray  
* index 개체에 .values 적용 >> ndarray 값 확인
* 단일 값 반환, 슬라이싱 가능
* 한 번 만들어진 인덱스 변경 불가
* 인덱스는 only 식별용 (연산에서 제외)
* .reset_index(): 데이터프레임과 seires 에 적용 / 새로운 인덱스를 연속형숫자로 할당 / 기존 인덱스는 'index' 칼럼으로 추가 >> 데이터프레임 반환 / 기존 인덱스가 연속 int 숫자형이 아닐 경우에 주로 사용 / drop=True ; 기존 인덱스 드롭하고 반환 >> 그대로 series

> **데이터 셀렉션 및 필터링**

(1) 데이터프레임 바로 뒤 [ ] 연산자
* [] 내에 가능한 것: 칼럼명, 칼럼명 리스트객체, 인덱스로 변환 가능한 표현식
   - 인덱스로 변환 가능한 표현식: [0:2] but 지양 / 불린 인덱싱 가능  

(2) 데이터프레임 iloc[] 연산자  
* iloc[]: 위치 기반 인덱싱 방식 /  행과 열 위치를, 0을 출발점으로 하는 세로축, 가로축 좌표 정숫값으로 지정 / 정수 슬라이싱 가능 / 인덱스값, 칼럼명 지정하면 오류
  - -1 사용: 맨 마지막 칼럼 *iloc[：, -1]* 또는 맨 마지막 칼럼만 제외 *iloc[:, ：-1]*
  - 불린 인덱싱 불가능

(3) 데이터프레임 loc[] 연산자
* loc[] : 명칭(label) 기반 인덱싱 방식 / loc[인덱스값, 칼럼명] 형식으로 데이터 추출
  - 슬라이싱 할 때, 종료값은 종료값
  - 불린 인덱싱 가능

(4) 불린 인덱싱
* 조건 여러개 일 때, 개별 조건은 () 사용
* 개별 조건을 새로운 변수로 할당 후, 이들을 사용해서 불린 인덱싱 수행 가능

> **정렬, Aggregation 함수, GroupBy 적용**

(1) 정렬 - sort_values()
* by= : 해당 칼럼 기준 정렬 수행
* ascending=True(디폴트) : 오름차순 / 내림차순
* inplace=False(디폴트) : 원본 유지, 정렬된 결과 반환

(2) Aggregation 함수 적용
* min(), max(), sum(), count() 등
* count() : NULL 제외
* 특정 칼럼 추출 후, aggregation 적용

(3) groupby() 적용
* by= : 해당 칼럼 기준으로 groupby
* DataFrameGroupBy 라는 별도의 데이터프레임을 반환
* groupby 적용 결과에 aggregation 적용 시, 그룹 칼럼을 **제외**한 모든 칼럼에 aggregation 적용
* groupby 후 특정 칼럼에 aggregation 적용 : DataFrameGroupBy에 해당 칼럼 필터링 후 함수 적용하면 된다.
* agg() : 적용하려는 aggregation 함수 여러 개 나열
   - 칼럼마다 서로 다른 함수 적용 : agg() 내에 딕셔너리 입력 (칼럼과 적용할 함수가 입력되어있는 딕셔너리)

> **결손 데이터 처리하기**

* 결손 데이터: 데이터값이 NULL
* *isna()* : 결손값 확인
* *fillna()* : 결손값을 다른 값으로 대체

(1) isna()로 결손 데이터 여부 확인
* dataframe.isna() : 모든 칼럼값에 대해 NaN 여부를 T/F 로 반환
* *isna().sum()* : 결손 데이터 개수 확인

(2) fillna()로 결손 데이터 대체
* 반환 값을 원래 칼럼에 정하거나 inplace=True >> 원본 데이터 변경됨

> **apply lambda 식으로 데이터 가공**

* apply 함수에 lambda 식 결합 > 레코드별 데이터 가공 / 복잡한 가공이 필요한 경우 사용
* lambda 식: 함수형 프로그래밍
   - def-retrun 같은 함수 선언과 함수 내 처리를 한 줄의 식으로 쉽게 변환하는 식
   - *lambda x ： x ** 2* (입력 인자:반환 계산식)
   - 여러 개의 값을 입릭 인자로 사용할 경우 *map()* 함수 결합
* 데이터프레임이 lambda 식 그대로 적용 가능
  - if else 절 사용 가능 (if 절에서는 식보다 반환 값을 먼저 적어야함 / else 절에서는 반환값이 나중)
  - if, else if, else 절 불가능
  - else if << else 절을 ()로 내포해 () 내에서 다시 if else 적용해야함  
  - else if 를 많이 반복해야 할 경우, 별도의 함수를 생성하는 것이 좋을 수 있음

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성.
def get_category(age):
  cat = ''
  if age <= 5: cat = 'Baby'
  elif age <= 12: cat = 'Child'
  elif age <= 18: cat = 'Teenager'
  elif age <= 25: cat = 'Student'
  elif age <= 35: cat = 'Young Adult'
  elif age <= 60: cat = 'Adult'
  else : cat = 'Elderly'

  return cat

# lambda 식에 위에서 생성한 get_category( ) 함수를 반환값으로 지정.
# get_category(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age','Age_cat']].head()
```
