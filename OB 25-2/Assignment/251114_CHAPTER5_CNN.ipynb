## CHAPTER 05 <합성곱 신경망 I>

## **5.1 합성곱 신경망**

✨ 복잡하고 시간이 오래 걸리는 딥러닝 문제를 해결

✨ 이미지 전체가 아닌 국소적 부분을 계산 (시간, 자원 절약)

>  ## 합성곱층의 필요성

* 이미지, 영상 처리에 유용: 원본 데이터의 구조를 펼칠 때 공간적 구조를 무시하게 되는 문제 방지 가능


>  ## 합성곱 신경망 구조  

◼ **합성곱 신경망(CNN)**: 음성 인식, 이미지, 영상 인식에서 주로 사용되는 신경망. 다차원 배열 데이터를 처리하도록 구성됨.

* (1) 입력층: 입력 이미지 데이터가 최초로 거치는 계층. 이미지는 3차원 데이터 (높이, 너비, 채널) (채널 = 1(그레이스케일) OR 3 (컬러))

* (2) 합성곱층: 입력 이미지의 주요 특성 추출.
* **특성 추출** : 커널/필터(주로 3x3, 5x5 / 스트라이드에 따라 순차적 이동) 사용. 커널이 스트라이드 간격만큼 순회하며 모든 입력값과의 합성곱 연산 ▶ 특성 맵 추출
* 💠ex) 스트라이드=1, 그레이스케일 일때 이동 과정

     -- 1. 입력이미지에 3x3 필터 적용: 이미와 필터 포갠 다음 대응되는 숫자끼리 곱해서 합계.

     -- 2. 필터가 1만큼 순차적으로 이동 (스트라이드)

     -- 3. 원본 (6,6,1) -> (4,4,1) 크기의 특성 맵

* 💠컬러이미지 일때의 합성곱: 필터 채널이 3. RGB 각각에 서로 다른 가중치로 합성곱을 적용한 후 결과를 더해줌. ❗필터개수는 1개

* 💠필터가 2개 이상일 때의 합성곱: 필터 각각은 특성 추출 결과의 채널이 됨.

* ✅ [입력 데이터] W1 * H1 * D1 / [하이퍼 파라미터] 필터 개수 K, 필터 크기 F, 스트라이드 S, 패딩 P / [출력 데이터] W2 = (W1-F+2P)÷S+1, H2 = (H1-F+2P)÷S+1, D2 = K

* (3) 풀링층: 특성 맵의 차원을 다운 샘플링(이미지를 축소하는것). 주요 특성 벡터 추출.

* **최대 풀링**: 대상 영역에서 최댓값 추출 (주로 사용)
* 평균 풀링: 대상 영역에서 평균을 반환

* 💠 최대 풀링 연산 과정: 스트라이드에 따라 이동하며 해당하는 영역 중 최댓값 선택 (반복)

* ✅ [입력 데이터] W1 * H1 * D1 / [하이퍼 파라미터] 필터 크기 F, 스트라이드 S / [출력 데이터] W2 = (W1-F)÷S+1, H2 = (H1-F)÷S+1, D2 = D1

* (4) 완전 연결층: 차원이 축소된 특성 맵이 전달됨. 주요 특성 벡터들을 1차원 벡터로 변환. (flattened)

* (5) 출력층: 입력받은 값을 소프트맥스 활성화 함수를 사용해 0~1 사이의 값으로 출력.


>  ## 1D, 2D, 3D 합성곱

* 이동하는 방향수, 출력 형태에 따라 분류

◼ **1D 합성곱**: 필터가 시간을 축으로 좌우로만 이동가능한 합성곱. 그래프 곡선을 완화할 때 사용.

* 입력 W + 필터 k ▶ 출력 W


◼ **2D 합성곱**: 필터가 2개의 방향으로 이동.

* 입력 (W,H) + 필터 (k,k) ▶ 출력 (W,H)


◼ **3D 합성곱**: 필터가 3개의 방향으로 이동.

* 입력 (W,H,L) + 필터 (k,k,d) ▶ 출력 (W,H,L)
* d < L 을 유지하는 것 중요

◼ **3D 입력을 갖는 2D 합성곱**: 입력이 3D 형태인데 출력 형태가 2D. 필터의 길이가 입력채널의 길이와 동일하기 때문.

* 입력 (W,H,L) + 필터 (k,k,L) ▶ 출력 (W,H)

◼ **1X1 합성곱**: 3D 형태로 입력

* 입력 (W,H,L) + 필터 (1,1,L) ▶ 출력 (W,H)
