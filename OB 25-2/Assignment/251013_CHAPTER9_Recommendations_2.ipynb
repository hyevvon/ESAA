````python
import pandas as pd
import numpy as np
import warnings; warnings.filterwarnings('ignore')

movies = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/tmdb_5000_movies.csv')
print(movies.shape)
movies.head(1)
````

````python
## Ï£ºÏöî ÏπºÎüº Ï∂îÏ∂ú
movies_df = movies[['id', 'title', 'genres', 'vote_average', 'vote_count', 'popularity',
                    'keywords', 'overview']]
````

````python
pd.set_option('max_colwidth', 100)
movies_df[['genres', 'keywords']][:1]
````

````python
from ast import literal_eval
movies_df['genres'] = movies_df['genres'].apply(literal_eval)
movies_df['keywords'] = movies_df['keywords'].apply(literal_eval)
````

````python
movies_df['genres'] = movies_df['genres'].apply(lambda x: [y['name'] for y in x])
movies_df['keywords'] = movies_df['keywords'].apply(lambda x: [y['name'] for y in x])
movies_df[['genres', 'keywords']][:1]
````

````python
from sklearn.feature_extraction.text import CountVectorizer

# CountVectorizer Ï†ÅÏö©ÏùÑ ÏúÑÌï¥ Í≥µÎ∞± Î¨∏ÏûêÎ°ú word Îã®ÏúÑÍ∞Ä Íµ¨Î∂ÑÎêòÎäî Î¨∏ÏûêÏó¥Î°ú Î≥ÄÌôò
movies_df['genres_literal'] = movies_df['genres'].apply(lambda x: (' ').join(x))
count_vect = CountVectorizer(min_df=0.0, ngram_range=(1, 2))
genre_mat = count_vect.fit_transform(movies_df['genres_literal'])
print(genre_mat.shape)
````

````python
from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(genre_mat, genre_mat)
print(genre_sim.shape)
print(genre_sim[:2])
````

* ‚úÖ genre_sim: movies_df Ïùò ÌñâÎ≥Ñ Ïû•Î•¥ Ïú†ÏÇ¨ÎèÑ Í∞íÏùÑ Í∞ÄÏßÄÍ≥† ÏûàÎäî Í∞ùÏ≤¥

````python
genre_sim_sorted_ind = genre_sim.argsort()[:, ::-1]
print(genre_sim_sorted_ind[:1])
````

````python
def find_sim_movie(df, sorted_ind, title_name, top_n=10): #Ï∂îÏ≤ú Í∏∞Ï§Ä ÏòÅÌôî, Ï∂îÏ≤úÌï† ÏòÅÌôî Ïàò

  # Ïù∏ÏûêÎ°ú ÏûÖÎ†•Îêú movies_df DataFrameÏóêÏÑú 'title' ÏπºÎüºÏù¥ ÏûÖÎ†•Îêú title_name Í∞íÏù∏ DataFrame Ï∂îÏ∂ú
  title_movie = df[df['title'] == title_name]

  # title_nameÏùÑ Í∞ÄÏßÑ DFÏùò index Í∞ùÏ≤¥Î•º ndarrayÎ°ú Î∞òÌôòÌïòÍ≥†
  # sorted_ind Ïù∏ÏûêÎ°ú ÏûÖÎ†•Îêú genre_sim_sorted_ind Í∞ùÏ≤¥ÏóêÏÑú Ïú†ÏÇ¨ÎèÑ ÏàúÏúºÎ°ú top_nÍ∞úÏùò index Ï∂îÏ∂ú
  title_index = title_movie.index.values
  similar_indexes = sorted_ind[title_index, :(top_n)]

  # Ï∂îÏ∂úÎêú top_n index Ï∂úÎ†•, top_n indexÎäî 2Ï∞®Ïõê Îç∞Ïù¥ÌÑ∞ÏûÑ
  # dataframeÏóêÏÑú indexÎ°ú ÏÇ¨Ïö©ÌïòÍ∏∞ ÏúÑÌï¥ÏÑú 1Ï∞®Ïõê arrayÎ°ú Î≥ÄÍ≤Ω
  print(similar_indexes)
  similar_indexes = similar_indexes.reshape(-1)

  return df.iloc[similar_indexes]
````

````python
percentile = 0.6
m = movies_df['vote_count'].quantile(percentile)
C = movies_df['vote_average'].mean()

def weighted_vote_average(record):
  v = record['vote_count']
  R = record['vote_average']

  return ((v/(v+m))*R)+((m/(m+v))*C)

movies_df['weighted_vote'] = movies.apply(weighted_vote_average, axis=1)
````

````python
def find_sim_movie(df, sorted_ind, title_name, top_n=10):
  title_movie = df[df['title']==title_name]
  title_index = title_movie.index.values

  # top_nÏùò 2Î∞∞Ïóê Ìï¥ÎãπÌïòÎäî Ïû•Î•¥ Ïú†ÏÇ¨ÏÑ±Ïù¥ ÎÜíÏùÄ Ïù∏Îç±Ïä§ Ï∂îÏ∂ú
  similar_indexes = sorted_ind[title_index, :(top_n*2)]
  similar_indexes = similar_indexes.reshape(-1)
  # Í∏∞Ï§Ä ÏòÅÌôî Ïù∏Îç±Ïä§Îäî Ï†úÏô∏
  similar_indexes = similar_indexes[similar_indexes != title_index]

  # top_nÏùò 2Î∞∞Ïóê Ìï¥ÎãπÌïòÎäî ÌõÑÎ≥¥Íµ∞ÏóêÏÑú weighted_voteÍ∞Ä ÎÜíÏùÄ ÏàúÏúºÎ°ú top_nÎßåÌÅº Ï∂îÏ∂ú
  return df.iloc[similar_indexes].sort_values('weighted_vote',ascending = False)[:top_n]

similar_movies = find_sim_movie(movies_df, genre_sim_sorted_ind, 'The Godfather',10)
similar_movies[['title','vote_average','weighted_vote']]
````

````python
import pandas as pd
import numpy as np

movies = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/movies.csv')
ratings = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/ratings.csv')
print(movies.shape)
print(ratings.shape)
````

````python
ratings = ratings[['userId','movieId','rating']]
ratings_matrix = ratings.pivot_table('rating',index='userId',columns='movieId')
ratings_matrix.head(3)
````

````python
# title ÏπºÎüºÏùÑ ÏñªÍ∏∞ ÏúÑÌï¥ moviesÏôÄ Ï°∞Ïù∏
ratings_movies = pd.merge(ratings, movies, on='movieId')

# columns = 'title'Î°ú title ÏπºÎüºÏúºÎ°ú ÌîºÎ≤ó ÏàòÌñâ
ratings_matrix = ratings_movies.pivot_table('rating',index='userId',columns='title')

# NaNÍ∞íÏùÑ Î™®Îëê 0ÏúºÎ°ú Î≥ÄÌôò
ratings_matrix = ratings_matrix.fillna(0)
ratings_matrix.head(3)
````

````python
# Transpose
ratings_matrix_T = ratings_matrix.transpose()
ratings_matrix_T.head(3)
````

````python
from sklearn.metrics.pairwise import cosine_similarity

item_sim = cosine_similarity(ratings_matrix_T, ratings_matrix_T)

# cosine_similarity()Î°ú Î∞òÌôòÎêú ÎÑòÌååÏù¥ ÌñâÎ†¨ÏùÑ ÏòÅÌôîÎ™ÖÏùÑ Îß§ÌïëÌï¥ DataFrameÏúºÎ°ú Î≥ÄÌôò
item_sim_df = pd.DataFrame(data=item_sim, index=ratings_matrix.columns,
                           columns = ratings_matrix.columns)

print(item_sim_df.shape)
item_sim_df.head(3)
````

````python
## ÎåÄÎ∂ÄÏôÄ Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÜíÏùÄ ÏÉÅÏúÑ 6Í∞ú Ï∂îÏ∂ú
item_sim_df['Godfather, The (1972)'].sort_values(ascending = False)[:6]
````

````python
item_sim_df['Inception (2010)'].sort_values(ascending = False)[:6]
````

````python
## ÏµúÏ†ÅÌôî ÌèâÏ†ê ÏòàÏ∏° Ìï®Ïàò
def predict_rating(ratings_arr, item_sim_arr):
  ratings_pred = ratings_arr.dot(item_sim_arr)/np.array([np.abs(item_sim_arr).sum(axis=1)])
  return ratings_pred
````

````python
ratings_pred = predict_rating(ratings_matrix.values, item_sim_df.values)
ratings_pred_matrix = pd.DataFrame(data = ratings_pred, index = ratings_matrix.index,
                                   columns = ratings_matrix.columns)
ratings_pred_matrix.head(3)
````

````python
## MSE ÌèâÍ∞Ä
from sklearn.metrics import mean_squared_error

# ÏÇ¨Ïö©ÏûêÍ∞Ä ÌèâÏ†êÏùÑ Î∂ÄÏó¨Ìïú ÏòÅÌôîÏóê ÎåÄÌï¥ÏÑúÎßå ÏòàÏ∏° ÏÑ±Îä• ÌèâÍ∞Ä MSEÎ•º Íµ¨Ìï®.
def get_mse(pred, actual):
  # ÌèâÏ†êÏù¥ ÏûàÎäî Ïã§Ï†ú ÏòÅÌôîÎßå Ï∂îÏ∂ú
  pred = pred[actual.nonzero()].flatten()
  actual = actual[actual.nonzero()].flatten()
  return mean_squared_error(pred, actual)

print('ÏïÑÏù¥ÌÖú Í∏∞Î∞ò Î™®Îì† ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ MSE:', get_mse(ratings_pred, ratings_matrix.values))
````

````python
def predict_rating_topsim(ratings_arr, item_sim_arr, n=20):
  # ÏÇ¨Ïö©Ïûê-ÏïÑÏù¥ÌÖú ÌèâÏ†ê ÌñâÎ†¨ ÌÅ¨Í∏∞ÎßåÌÅº 0ÏúºÎ°ú Ï±ÑÏö¥ ÏòàÏ∏° ÌñâÎ†¨ Ï¥àÍ∏∞Ìôî
  pred = np.zeros(ratings_arr.shape)

  # ÏÇ¨Ïö©Ïûê-ÏïÑÏù¥ÌÖú ÌèâÏ†ê ÌñâÎ†¨Ïùò Ïó¥ ÌÅ¨Í∏∞ÎßåÌÅº Î£®ÌîÑ ÏàòÌñâ
  for col in range(ratings_arr.shape[1]):
    # Ïú†ÏÇ¨ÎèÑ ÌñâÎ†¨ÏóêÏÑú Ïú†ÏÇ¨ÎèÑÍ∞Ä ÌÅ∞ ÏàúÏúºÎ°ú nÍ∞ú Îç∞Ïù¥ÌÑ∞ ÌñâÎ†¨Ïùò Ïù∏Îç±Ïä§ Î∞òÌôò
    top_n_items = [np.argsort(item_sim_arr[:,col])[:-n-1:-1]]
    # Í∞úÏù∏ÌôîÎêú ÏòàÏ∏° ÌèâÏ†êÏùÑ Í≥ÑÏÇ∞
    for row in range(ratings_arr.shape[0]):
      pred[row, col] = item_sim_arr[col,:][top_n_items].dot(ratings_arr[row, :][top_n_items].T)
      pred[row, col] /= np.sum(np.abs(item_sim_arr[col,:][top_n_items]))

  return pred
````

````python
ratings_pred = predict_rating_topsim(ratings_matrix.values, item_sim_df.values, n=20)
print('ÏïÑÏù¥ÌÖú Í∏∞Î∞ò ÏµúÍ∑ºÏ†ë Top-20 Ïù¥ÏõÉ MSE: ', get_mse(ratings_pred, ratings_matrix.values))

# Í≥ÑÏÇ∞Îêú ÏòàÏ∏° ÌèâÏ†ê Îç∞Ïù¥ÌÑ∞Îäî DataFrameÏúºÎ°ú Ïû¨ÏÉùÏÑ±
ratings_pred_matrix = pd.DataFrame(data=ratings_pred, index=ratings_matrix.index,
                                   columns=ratings_matrix.columns)
````

````python
def get_unseen_movies(ratings_matrix, userId):
    # userIdÎ°ú ÏûÖÎ†•Î∞õÏùÄ ÏÇ¨Ïö©ÏûêÏùò Î™®Îì† ÏòÅÌôî Ï†ïÎ≥¥Î•º Ï∂îÏ∂úÌï¥ SeriesÎ°ú Î∞òÌôòÌï®.
    # Î∞òÌôòÎêú user_ratingÏùÄ ÏòÅÌôîÎ™Ö(title)ÏùÑ Ïù∏Îç±Ïä§Î•º Í∞ÄÏßÄÎäî Series Í∞ùÏ≤¥ÏûÑ.
    user_rating = ratings_matrix.loc[userId, :]

    #user_ratingÏù¥ 0Î≥¥Îã§ ÌÅ¨Î©¥ Í∏∞Ï°¥Ïóê Í¥ÄÎûåÌïú ÏòÅÌôîÏûÑ. ÎåÄÏÉÅ Ïù∏Îç±Ïä§Î•º Ï∂îÏ∂úÌï¥ list Í∞ùÏ≤¥Î°ú ÎßåÎì¶.
    already_seen = user_rating[ user_rating > 0].index.tolist()

    # Î™®Îì† ÏòÅÌôîÎ™ÖÏùÑ list Í∞ùÏ≤¥Î°ú ÎßåÎì¶.
    movies_list = ratings_matrix.columns.tolist()

    # list comporehensionÏúºÎ°ú already_seenÏóê Ìï¥ÎãπÌïòÎäî ÏòÅÌôîÎäî movies_listÏóêÏÑú Ï†úÏô∏Ìï®.
    unseen_list = [ movie for movie in movies_list if movie not in already_seen]

    return unseen_list
````

````python
def recomm_movie_by_userid(pred_df, userId, unseen_list, top_n = 10):
    # ÏòàÏ∏° ÌèâÏ†ê DataFrameÏóêÏÑú ÏÇ¨Ïö©Ïûêid Ïù∏Îç±Ïä§ÏôÄ unseen_listÎ°ú Îì§Ïñ¥Ïò® ÏòÅÌôîÎ™Ö ÏπºÎüºÏùÑ Ï∂îÏ∂úÌï¥
    # Í∞ÄÏû• ÏòàÏ∏° ÌèâÏ†êÏù¥ ÎÜíÏùÄ ÏàúÏúºÎ°ú Ï†ïÎ†¨Ìï®.
    recomm_movies = pred_df.loc[userId, unseen_list].sort_values(ascending=False)[:top_n]
    return recomm_movies

# ÏÇ¨Ïö©ÏûêÍ∞Ä Í¥ÄÎûåÌïòÏßÄ ÏïäÏùÄ ÏòÅÌôîÎ™Ö Ï∂îÏ∂ú
unseen_list = get_unseen_movies(ratings_matrix, 9)

# ÏïÑÏù¥ÌÖú Í∏∞Î∞òÏùò ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ ÌòëÏóÖ ÌïÑÌÑ∞ÎßÅÏúºÎ°ú ÏòÅÌôî Ï∂îÏ≤ú
recomm_movies = recomm_movie_by_userid(ratings_pred_matrix, 9, unseen_list, top_n=10)

# ÌèâÏ†ê Îç∞Ïù¥ÌÑ∞Î•º DataFrameÏúºÎ°ú ÏÉùÏÑ±.
recomm_movies = pd.DataFrame(data=recomm_movies.values, index=recomm_movies.index,
                             columns=['pred_score'])

recomm_movies
````
